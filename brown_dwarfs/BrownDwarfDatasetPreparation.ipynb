Find all the TICs who's period matches the reported period of the 

```python
len_total = 0;
for id, period in zip(list, periods):
    print(str(df[(df["target_id"] == id) & (abs(df["tce_period"] - period) < 0.1)]['tce_period']) + " "+ str(period) +" "+ str(id))
    len_total += len(df[(df["target_id"] == id) & (abs(df["tce_period"] - period) < 0.1)]["target_id"])
```

```python
diffs = []
for id, period in zip(list, periods):
    curr_periods = df[(df["target_id"] == id) & (abs(df["tce_period"] - period) < 0.1)]["tce_period"]
    curr_depths = df[(df["target_id"] == id) & (abs(df["tce_period"] - period) < 0.1)]["tce_depth"]
    curr_durations = df[(df["target_id"] == id) & (abs(df["tce_period"] - period) < 0.1)]["tce_duration"]
    print(np.std(curr_periods)/np.mean(curr_periods))
    print(np.std(curr_depths)/np.mean(curr_depths))
    print(np.std(curr_durations)/np.mean(curr_durations))
    print(df[(df["target_id"] == id) & (abs(df["tce_period"] - period) < 0.1)][["tce_period","tce_depth","tce_duration"]]);
    print("---------------------------------------")
```

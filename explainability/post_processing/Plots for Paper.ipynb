```python
import matplotlib.pyplot as plt
import numpy as np
%pylab inline
```

```python
prec_new_exp=np.load('/nobackup/khauskne/kdd/prec_thresh_new_exp.npy')
rec_new_exp=np.load('/nobackup/khauskne/kdd/rec_thresh_new_exp.npy')

prec_occl=np.load('/nobackup/khauskne/kdd/prec_thresh_occl.npy')
rec_occl=np.load('/nobackup/khauskne/kdd/rec_thresh_occl.npy')

prec_simpleshap=np.load('/nobackup/khauskne/kdd/prec_thresh_simpleshap.npy')
rec_simpleshap=np.load('/nobackup/khauskne/kdd/rec_thresh_simpleshap.npy')
```

```python
plt.figure(figsize=(8,5))
plt.plot(prec_new_exp[0], prec_new_exp[1], color='r', label='New Technique')
plt.plot(prec_occl[0], prec_occl[1], color='c', label='Occlusion')
plt.plot(prec_simpleshap[0], prec_simpleshap[1], color='b', label='Simple Shap')
plt.legend()
plt.title('Evaluating Precision of Explainability Techniques Based on Threshold')
plt.xlabel('Threshold Used')
plt.ylabel('Precision')
plt.savefig('threshold_prec_comparison.pdf')
```

```python
plt.figure(figsize=(8,5))
plt.plot(rec_new_exp[0], rec_new_exp[1], color='r', label='New Technique')
plt.plot(rec_occl[0], rec_occl[1], color='c', label='Occlusion')
plt.plot(rec_simpleshap[0], rec_simpleshap[1], color='b', label='Simple Shap')
plt.legend()
plt.title('Evaluating Recall of Explainability Techniques Based on Threshold')
plt.xlabel('Threshold Used')
plt.ylabel('Recall')
plt.savefig('threshold_recall_comparison.pdf')
```

```python
plt.figure(figsize=(8,5))
plt.plot(prec_new_exp[0], prec_new_exp[1], color='r', label='New Technique Precision')
plt.plot(rec_new_exp[0], rec_new_exp[1], color='r', linestyle='--', label='New Technique Recall' )
plt.plot(prec_occl[0], prec_occl[1], color='c', label='Occlusion Precision')
plt.plot(rec_occl[0], rec_occl[1], color='c', linestyle='--', label='Occlusion Recall')
plt.plot(prec_simpleshap[0], prec_simpleshap[1], color='b', label='Simple Shap Precision')
plt.plot(rec_simpleshap[0], rec_simpleshap[1], color='b', linestyle='--', label='Simple Shap Recall')
plt.legend(loc=(1.01,0.5))
plt.title('Evaluating Precision and Recall of Explainability Techniques Based on Threshold')
plt.xlabel('Threshold Used')
plt.ylabel('Precision/Recall')
plt.savefig('threshold_prec_and_rec_comparison.pdf')
```
